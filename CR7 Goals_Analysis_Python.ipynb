```python
#Analysis
#Goals by competition
goals_by_competion = df.groupby('Competition')['Season'].count()
goals_by_competion
```

```python
#Goals by Venue
goals_by_Venue = df.groupby('Venue')['Venue'].count()

goals_by_Venue
```

```python
#Goals by Season
goals_by_Season = df.groupby('Season')['Season'].count()

goals_by_Season
```

```python
#Goals by Clubs 
goals_by_Clubs = df.groupby('Club')['Club'].count()

goals_by_Clubs
```

```python
#Goals by Type
goals_by_Type = df.groupby('Type')['Type'].count()

goals_by_Type
```

```python
#Goals by goal_assist
goals_by_goal_assist= df.groupby('Goal_assist')['Goal_assist'].count()

goals_by_goal_assist

```

```python
#goals for realmadrid
goals_for_realmadrid = df[df['Club']=='Real Madrid']['Club'].count()

goals_for_realmadrid

```

```python
#visualizations
sns.countplot(x='Venue',data=df,color='brown')
plt.title('Goals by Venue')
plt.xlabel('Venue')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Club',data=df)
plt.title('Goals by Clubs')
plt.xlabel('Club')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Club',data=df,hue='Venue')
plt.title('Goals by Clubs')
plt.xlabel('Club')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x='Season',data=df,color='pink')
plt.title('Goals Over Seasons')
plt.xlabel('Season')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x='Season',data=df,hue='Venue',color='pink',palette='coolwarm')
plt.title('Goals Over Seasons')
plt.xlabel('Season')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(20,4))
sns.countplot(x='Type',data=df,color='red')
plt.title('Goals by Type')
plt.xlabel('Type')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(20,4))
sns.countplot(x='Type',data=df,hue='Venue',palette='deep')
plt.title('Goals by Type')
plt.xlabel('Type')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(22,4))
sns.countplot(x='Competition',data=df,color='grey')
plt.title('Goals by Competition')
plt.xlabel('Competition')
plt.ylabel('Count of Goals')
```

```python
plt.figure(figsize=(22,4))
sns.countplot(x='Competition',data=df,hue='Venue',palette='viridis')
plt.title('Goals by Competition')
plt.xlabel('Competition')
plt.ylabel('Count of Goals')
```

```python
# Create a countplot for Playing Position
sns.countplot(data=df, x='Playing_Position', color='orange')

plt.title('Goals by Playing Position')
plt.xlabel('Playing Position')
plt.ylabel('Count of Goals')
plt.show()
```

```python
# Create a countplot for Playing Position
sns.countplot(data=df, x='Playing_Position',hue='Venue', palette='coolwarm')

plt.title('Goals by Playing Position')
plt.xlabel('Playing Position')
plt.ylabel('Count of Goals')
plt.show()
```
